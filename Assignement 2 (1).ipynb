```python
import pandas as pd
```

```python
import sqlite3

# Connecting to sqlite
# connection object
connection_obj = sqlite3.connect('sales.db')

# cursor object
cursor_obj = connection_obj.cursor()

# Drop the GEEK table if already exists.
cursor_obj.execute("DROP TABLE IF EXISTS SALES")

# Creating table
table = """ CREATE TABLE SALES (Email VARCHAR(255) NOT NULL,Phone INT,DateOfPurchase DATE,SKUs bought REAL,TotalPrice REAL,Discount REAL); """

cursor_obj.execute(table)

print("Table is Ready")

# Close the connection
connection_obj.close()
```
